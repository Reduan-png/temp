```python
#13(b) attempt 1 part 1

import numpy as np
import matplotlib.pyplot as plt
from sympy import Symbol, zeta
from sympy.utilities import lambdify

# Defining s and zeta_exp
s = Symbol('s')
zeta_exp = zeta(s)

# Subsequently converting zeta_exp into a numerical function
zeta_fnc = lambdify(s, zeta_exp)

# Generating values for s with a negative array to plot to the right of the asymptote at 1
s_x = np.linspace(-1.99, -0.01, 100)

# Calculating the corresponding zeta(s) values and storing into zeta_y
zeta_y = zeta_fnc(s_x)

# Ploting the graph using semilogy to have log scaling on the y-axis
plt.semilogy(1 - s_x, np.abs(zeta_y))
plt.xlabel("1 - s")
plt.ylabel("|ζ(s)|")
plt.title("Graph of ζ(s) against 1 - s")
# Including grid to increase clarity
plt.grid(True)
plt.show()
```

```python
#13(c) part 1

import numpy as np
import matplotlib.pyplot as plt

# Importing Symbol for variable assignment, sympify to help define rational numbers, whilst zeta and prime are neccessary
from sympy import Symbol, S, zeta, prime

# Defining the symbolic variables 's'  and 'p'
s = Symbol('s')
p = Symbol('p')

# Defining the function for the Euler product term (1 - p^(-4))^(-1) wherein I utilise prime from sympy
def euler_product(n):
    prime_set = [prime(i) for i in range(1, n+1)]
    final_product = [(1 - (S(1)/p**4)**(-1)) for p in prime_set]
    return np.prod(final_product)

# Generating prime numbers up to 100
prime_numbers = []
i = 2
while len(prime_numbers) < 100:
    if prime(i):
        prime_numbers.append(i)
    i += 1

# Calculating the absolute difference for each prime number
diff_values = []
for p in prime_numbers:
    diff_values.append(abs(zeta(4) - euler_product(p)))

# Plotting the graph
plt.plot(prime_numbers, diff_values)
plt.xlabel("Prime Number (p)")
plt.ylabel("Absolute Difference")
plt.title("Absolute Difference between ζ(4) and Euler's Product")
plt.grid(True)
plt.show()
```

```python
#14(a)

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma
from math import factorial

# So that I can plot them both on the same graph
%matplotlib

# Generating values on the domain [-5, 5]
x = np.linspace(-5, 5, 1000)

# Calculating gamma values for each x
gamma_y = gamma(x)

# Plotting the graph of Gamma
plt.plot(x, gamma_y)
plt.xlabel("x")
plt.ylabel("Gamma(x)")
plt.title("Gamma Function on [-5, 5]")
plt.grid(True)
plt.show()

# Generating integers on the domain [-5, 5]
x = np.arange(-5, 6)

# Calculating the (x-1)! for each integer x
factorial_y = [factorial(xi-1) if xi > 0 else float('nan') for xi in x]

# Plotting the graph of (x-1)!
plt.plot(x, factorial_y, marker='o', linestyle='-', color='blue')
plt.xlabel("x")
plt.ylabel("(x-1)!")
plt.title("Factorial Function (x-1)! for Integer Values")
plt.grid(True)
plt.show()
```

```python
#14(d). part 1

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma
from math import sqrt, pi, exp

# Defining Stirling's approximation
def stirling_approximation(x):
    return sqrt(2*pi*x) * (x**(x-0.5)) * exp(-x)

x_inputs = np.linspace(1, 10, 100)

# Calculating the gamma values and error with a for loop
gamma_values = []
fractional_error = []

for xi in x_inputs:
    if xi > 0:
        gamma_val = gamma(xi)
        gamma_values.append(gamma_val)
        fractional_err = abs((stirling_approximation(xi) - gamma_val) / gamma_val)
        fractional_error.append(fractional_err)

# Plotting the error between the approximation and gamma
plt.plot(x_inputs[x_inputs > 0], fractional_error)
plt.xlabel("x")
plt.ylabel("Fractional Error")
plt.title("The Fractional Error between Stirling's Approximation and Γ(x)")
plt.grid(True)
plt.show()
```

```python
#14(d). part 2

import numpy as np
from scipy.special import gamma
from math import sqrt, pi, exp

# Defining Stirling's approximation function
def stirling_approximation(x):
    return sqrt(2*pi*x) * (x**(x-0.5)) * exp(-x)

# Defining the function to check for our desired level accuracy
def check_level_of_accuracy(x):
    gamma_val = gamma(x)
    approx_val = stirling_approximation(x)
    fractional_err = abs((approx_val - gamma_val) / gamma_val)
    return fractional_err < 0.01

x_inputs = np.linspace(0.05, 100.05, 1000)

accurate_inputs = [x for x in x_inputs if check_level_of_accuracy(x)]

print("Values of x that have percent-level accuracy:", accurate_inputs)
```

```python
# 17(d)

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Defining the function for the Mobius transformation
def mobius_transform(a, b, c, d, z):
    return (a * z + b) / (c * z + d)

# Defining the parameter values and hence the transformation
a, b, c, d = 1, -1j/2, 1j/2, 1
# a, b, c, d = 0, 1, 1, 0
# a, b, c, d = 1, 1j, 1, -1j

# Generating input values
theta = np.linspace(0, 2*np.pi, 100)
R_inputs = np.linspace(0, 1, 10)

# Ploting the images of the circles
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for R in R_inputs:
    z = R * np.exp(1j * theta)
    w = mobius_transform(a, b, c, d, z)
    ax.plot(np.real(w), np.imag(w), R*np.ones_like(w), label=f'R={R}')

ax.set_xlabel('Real(z)')
ax.set_ylabel('Imaginary(z)')
# R represents the radius
ax.set_zlabel('R')
ax.set_title('The Images of Circles under Mobius Transformation')
ax.grid(True)

# Show the plot
plt.show()
```
